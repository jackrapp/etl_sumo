```python
url = "https://en.wikipedia.org/wiki/List_of_sumo_stables"
```

```python
tables = pd.read_html(url)
tables
```

```python
wiki_df = tables[0]
#wiki_df = wiki_df.columns[0:1]
#wiki_df.columns = ['heya', 'ichimon']
#df.drop(columns=[0, 1])
#wiki_d
```

```python
wiki = wiki_df.iloc[:,[0,1]]
wiki
```

```python
wiki.columns = ['Heya', 'Ichimon']
wiki = wiki.iloc[1:]
```

```python
wiki.Heya = wiki.Heya.replace({"Ōnomatsu": "Onomatsu", "Ōtake": "Otake"})
```

```python
finaldf = df.merge(wiki, how='outer' , left_on='heya', right_on='Heya')
```

```python
group_final = finaldf.groupby('heya')

owa = group_final.get_group('Nishiiwa')
```

```python
finaldf = finaldf.drop(columns='Heya')
```

```python
finaldf.to_sql('sumo', con=engine)
```
